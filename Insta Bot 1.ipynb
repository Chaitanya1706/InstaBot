### Q(6.1) Extract Username Of 500 Followers

```python
def openFollowersList(s):
    search = driver.find_element_by_xpath('//div[starts-with(@class,"LWmhU")]/input')
    search.send_keys(s)
    time.sleep(3)
    driver.find_element_by_partial_link_text(s).click()
    time.sleep(3)
    driver.find_element_by_partial_link_text('followers').click()
```

```python
def scroll():
    current_height= driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')
    for i in range(100):
        driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0, arguments[0]);',current_height) 
        time.sleep(2)
        new_height = driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')
        if new_height == current_height:
            break;
        current_height = new_height
```

```python
def getFollowers(profile, n):
    openFollowersList(profile)
    time.sleep(3)
    scroll()
    time.sleep(2)
    i=0
    user = driver.find_elements_by_xpath('//span[starts-with(@class,"Jv7Aj")]/a')
    followers = []
    for u in user:
        followers.append(u.text)
        i += 1
        if i == n:
            break
    btnClose = driver.find_elements_by_class_name('WaOAr')[-1].find_element_by_xpath('button[contains(@class, "wpO6b")]').click()
    return followers
```

```python
# 500 followers of 'foodtalkindia'
followers = getFollowers('foodtalkindia',500)
print(len(followers))
for f in followers:
    print(f)
```

```python
# 500 followers of 'So Delhi'
followers = getFollowers('So Delhi',500)
print(len(followers))
for f in followers:
    print(f)
```

### Q(6.2) -> Extract List of Mutual Followers of 'foodtalkindia' 

```python
search = driver.find_element_by_xpath('//div[starts-with(@class,"LWmhU")]/input')
search.send_keys('foodtalkindia')
time.sleep(3)
driver.find_element_by_partial_link_text('foodtalkindia').click()
time.sleep(2)
driver.find_element_by_partial_link_text('Followed by').click()
time.sleep(2)

#scroll down the list
current_height= driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')
while True:
    driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0, arguments[0]);',current_height) 
    time.sleep(1)
    new_height = driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')
    if new_height == current_height:
        break;
    current_height = new_height

#get the list of mutual followers
user = driver.find_elements_by_xpath('//span[starts-with(@class,"Jv7Aj")]/a')
mutual = []
for u in user:
    mutual.append(u.text)

btnClose = driver.find_elements_by_class_name('WaOAr')[-1].find_element_by_xpath('button[contains(@class, "wpO6b")]').click()
time.sleep(1)
if len(mutual) == 0:
    print("You don't have any mutual")
else:
    #open my profile
    profile = driver.find_elements_by_xpath("//div[@class = '_47KiJ']/div")
    profile[4].click()
    driver.find_element_by_partial_link_text('Profile').click()
    time.sleep(2)
    
    #extracting list of my followers
    driver.find_element_by_partial_link_text('followers').click()
    time.sleep(3)
    scroll()
    time.sleep(2)
    user = driver.find_elements_by_xpath('//span[starts-with(@class,"Jv7Aj")]/a')
    followers = []
    for u in user:
        followers.append(u.text)

    btnClose = driver.find_elements_by_class_name('WaOAr')[-1].find_element_by_xpath('button[contains(@class, "wpO6b")]').click()
    
    #finding out mutual follower who do not follow me back
    final_list = []
    for m in mutual:
        if m not in followers:
            final_list.append(m)
    if len(final_list) == 0:
        print('Every One Following Back')
    else:
        for i in final_list:
            print(i)
```
